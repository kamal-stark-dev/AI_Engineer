### Stream
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.

As we can see the response above took 12 seconds to generate, so streaming here can make the user's experience better.

````python
# basic text streaming
for chunk in llm.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="|", flush=True)
````

````python
# stream tool calls, reasoning, and other content

for chunk in llm.stream("What color is the sky?"):
    for block in chunk.content_blocks:
        if block["type"] == "reasoning" and (reasoning := block.get("reasoning")):
            print(f"Reasoning: {reasoning}")
        elif block["type"] == "tool_calling_chunk":
            print(f"Tool call chunk: {block}")
        elif block["type"] == "text":
            print(block["text"])
        else:
            pass
````

### Batch
Batching a collection of independent requests to a model can signifcantly improve performace and reduce costs, as the processing can be done in parallel.

````python
responses = llm.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])

for response in responses:
    print(responses.content)
````

By default, `batch()` will only return the final output for the entire batch. If you want to receive the output for each individual input as it finishes generating, you can stream the results iwth `batch_as_completed()`

> Note: when using `batch_as_completed()`, results may arrive out of order. Each includes the input index for matching to reconstruct the original order as needed.

````python
for response in llm.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
]):
    print(response.content)
````

When processing a large number of inputs using `batch()` or `batch_as_completed()`, you might want to limit the maximum number of parallel calls. This can be done by setting `max_concurrency` attribute in the `RunnableConfig` dictionary.

```py
llm.batch(
    list_of_inputs, 
    config={
        "max-concurrency": 5, # limit to 5 parallel calls
    }
)
```

### Runnable Chains (What actually makes Lang(Chain))

````python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
````

````python
from langchain_core.output_parsers import StrOutputParser

movie_title_template = PromptTemplate.from_template("Suggest me the title of a movie which of of the genre {genre} and language {language}, give me just the title of the single movie")

movie_title_chain = movie_title_template | llm | StrOutputParser()
````

````python
movie_title_chain.invoke({"genre": "Sci-Fi Year 2014", "language": "English"})
````

````python
# composed chain -> suppose you want to get summary of the movie title from another llm you can do the following

from langchain_core.runnables import RunnableLambda

movie_summary_prompt = PromptTemplate.from_template("Give me a 2-3 line summary of the movie {movie_title}")

def print_title(x):
    print(x["movie_title"])
    return x

composed_chain = (
    {"movie_title": movie_title_chain}
    | RunnableLambda(print_title)
    | movie_summary_prompt
    | llm
    | StrOutputParser()
)
````

````python
composed_chain.invoke({"genre": "Sci-Fi Year 2014", "language": "English"})
````

### Types of Chaining

1. **Sequential Chaining**: Also called extended chaining, here one task is followed by other task in a sequence like in the example above.
2. **Parallel Chaining**: A single piece of input data is simultaneously send to multiple AI agents/models to perform different independent tasks.
example: if you want to translate the movie summary generated above into multiple languages then you can do parallel chaining (as they are independent tasks.)
3. **Conditional Chaining**: Some task will happen if a condition is true.
example: an sentiment analyzer LLM returned "positive" then maybe generate a praise/thankyou email, and if it returned "negative" then maybe alert the manager about it.

````python
# instead of using the '|' operator we can also use `RunnableSequence`

from langchain_core.runnables import RunnableSequence

composed_chain = RunnableSequence({"movie_title": movie_title_chain} , (lambda x: print(x["movie_title"])) , movie_summary_prompt , llm , StrOutputParser())
````

````python
composed_chain.invoke({"language": "Hindi", "genre": "Court Room Drama"})
````

### Parallel Chaining (RunnableParallel)

````python
from langchain_core.runnables import RunnableParallel

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

joke_chain = (
    ChatPromptTemplate.from_template("tell me a joke about {topic}")
    | model
    | StrOutputParser()
)

poem_chain = RunnableSequence(
    ChatPromptTemplate.from_template("Write a 2-3 line poem about {topic}"),
    model,
    StrOutputParser()
)

runnable = RunnableParallel(joke=joke_chain, poem=poem_chain)

# Display Responses
runnable.invoke({"topic": "banana"})
````
